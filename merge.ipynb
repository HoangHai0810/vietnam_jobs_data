```python
import pandas as pd
```

```python
df_joboko = pd.read_csv("./pipelines/Crawler/JBK/Data/jobs/job_data.csv", encoding='utf-8', on_bad_lines='skip')
df_vietnamwork = pd.read_csv("./pipelines/Crawler/VNW/Data/jobs/vietnamwork.csv", encoding='utf-8', on_bad_lines='skip')
df_careerviet = pd.read_csv("./pipelines/Crawler/CRV/Data/jobs/careerviet.csv", encoding='utf-8', on_bad_lines='skip')
```

id, job_title, salary, due_date (careerviet, vnw, jbk = crawl(ing)_date + deadline), views (vietnamwork), skills (joboko: keywords, careerviet k co), posted_date (joboko k co, careerviet: created_date), level, category(vnw, jbk: field, crv: job_categories), min_yoe(jbk: profession, crv: experience), benefits (crv: job_benefits), address (crv: job_address, vnw, jbk: location), region(crv: location), company_name (crv: company), company_size (crv: k co)

```python
columns = ['job_title','salary','due_date','views','skills','posted_date','level','category','min_yoe','benefits','address','region','company_name','company_size']

df_vietnamwork = df_vietnamwork[['job_title','salary', 'crawling_date','deadline','views','skills','posted_date','level','field','min_yoe','benefits','location','region','company_name','company_size']]

df_vietnamwork.rename(columns={
    'field': 'category',
    'location': 'address',
}, inplace=True)
df_vietnamwork['source'] = 'vietnamwork'
```

```python
df_joboko = df_joboko[['job_title','salary', 'crawl_date','deadline','keyworks','level','field','profession','benefits','location','region','company_name','company_size']]

df_joboko.rename(columns={
    'keyworks': 'skills',
    'profession': 'min_yoe',
    'field': 'category',
    'location': 'address',
    'crawl_date': 'crawling_date',
}, inplace=True)
df_joboko['source'] = 'joboko'
```

```python
df_careerviet = df_careerviet[['job_title', 'salary', 'due_date','created_date','level','job_categories','experience','job_benefits','job_address','location','company']]

df_careerviet.rename(columns={
    'job_categories': 'category',
    'experience': 'min_yoe',
    'job_benefits': 'benefits',
    'job_address': 'address',
    'created_date': 'posted_date',
    'company': 'company_name',
    'location': 'region'
}, inplace=True)
df_careerviet['source'] = 'careerviet'
```

```python
# Merge các DataFrame lại với nhau
df_merged = pd.concat([df_joboko, df_vietnamwork, df_careerviet], ignore_index=True)

# Hiển thị thông tin của DataFrame đã merge
df_merged.info()
```

```python
df_merged.to_csv("./data/job_data_merged.csv", index=False)
```
